# Modelo regresion lineal con variables categoricas o Dummy

```python
# Transformaci√≥n de variables cualitativas (top 5 paises y artistas) a variables falsas/binarias o dummy
pd.options.display.max_columns = None

# Vamos a identificar las top 5 nacionalidades y artistas por frecuencia
top_nacionalidades = df['Nationality'].value_counts().head(3).index
top_artistas = df['Artist (Ind.)'].value_counts().head(3).index

# Ahora crearemos columnas binarias para estas top nacionalidades y artistas
for nationality in top_nacionalidades:
    df[f'Nationality_{nationality}'] = (df['Nationality'] == nationality).astype(int)

for artist in top_artistas:
    df[f'Artist_{artist}'] = (df['Artist (Ind.)'] == artist).astype(int)

# Muestra del dataframe con las nuevas columnas binarias
df.head(10)
```

```python
# modelo regresion con variables binarias top 5

import statsmodels.api as sm

# Preparar las variables independientes (X) y la variable dependiente (y)
X = df[[
    'Nationality_United States', 'Nationality_United Kingdom', 'Nationality_Puerto Rico',
    'Nationality_Canada', 'Nationality_Colombia','Nationality_Spain','Nationality_Brazil', 'Nationality_Australia','Nationality_South Korea',
    'Nationality_France','Artist_Bad Bunny','Artist_Ed Sheeran', 'Artist_Post Malone', 'Artist_J Balvin', 'Artist_The Weeknd', 'Artist_Drake', 'Artist_Billie Eilish',
    'Artist_Ozuna', 'Artist_XXXTENTACION', 'Artist_Dua Lipa'
]]
y = df['Points (Ind for each Artist/Nat)']


# Agregar una constante a las variables independientes
X = sm.add_constant(X)

# Construir el modelo
model = sm.OLS(y, X).fit()

# Mostrar el resumen del modelo
model_summary = model.summary()
model_summary
```

```python
# modelo regresion lineal

df= canciones[['Danceability', 'Energy', 'Loudness',
       'Speechiness', 'Acousticness', 'Instrumentalness', 'Valence', 'Points (Total)', 'Points (Ind for each Artist/Nat)']]

from sklearn.model_selection import train_test_split
x = df.drop(["Points (Ind for each Artist/Nat)"] , axis = 1).values
y = df["Points (Total)"].values
x_train , x_test , y_train ,y_test = train_test_split(x,y , test_size= 0.2 , random_state= 69)

from sklearn.preprocessing import RobustScaler
ro = RobustScaler()
x_train = ro.fit_transform(x_train)
x_test = ro.fit_transform(x_test)

from sklearn.linear_model import LinearRegression
ln = LinearRegression()
ln.fit(x_train,y_train)
y_pred = ln.predict(x_test)

from sklearn.metrics import mean_absolute_error, r2_score
mae = mean_absolute_error(y_test , y_pred)
mae
```

```python
reg_score = r2_score(y_test , y_pred)
reg_score
```

# Analisis con series de tiempo

```python
from pandas.plotting import scatter_matrix
import missingno as mano
import matplotlib.pyplot as plt
from datetime import datetime

dfa=canciones
dfa['Date']= pd.to_datetime(dfa['Date'])
dfa = dfa.set_index('Date')
dfa['Year']= dfa.index.year
dfa['Month'] = dfa.index.month
# transformaci√≥n formato object a formato fecha
```

```python
# series de tiempo para variables de interes (Danceability,Energy, Speechiness, Acousticness, Valence)
list_to_plot = ['Danceability','Energy','Speechiness','Acousticness','Valence']
for i in list_to_plot:
  fit, ax = plt.subplots(figsize=(20,7))
  sns.lineplot(data = dfa, y=i, x = 'Date', linewidth = 0.8)
  plt.title('Serie de tiempo de '+i)
  plt.show()
```

```python
# representaci√≥n de series de tiempo con grafico de densidad
for i in list_to_plot:
  fig, ax = plt.subplots(figsize=(14,7), )
  sns.kdeplot(data = dfa, x=i, hue='Year', fill=True, alpha=.5, linewidth=0, palette='viridis')
  plt.title('Representaci√≥n de serie de tiempo como densidad de '+i)
  plt.show()
```

```python
# Filtra las filas donde la nacionalidad es Estados Unidos
df_usa = canciones[canciones['Nationality'] == 'United States']

# Agrupa por fecha y cuenta las ocurrencias
dfus = df_usa.groupby('Date').size()

# Ahora time_series es tu serie de tiempo
print(dfus)
fit, ax = plt.subplots(figsize=(20,7))
sns.lineplot(data = dfus,y=dfus, x = 'Date', linewidth = 0.8)
plt.title('Serie de tiempo de cantidad USA')
plt.show()
```
